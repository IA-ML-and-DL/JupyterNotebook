# S1C03 - Estilos de las gráficas

````python
import numpy as np

np.random.seed(123)

x = np.arange(0,100,1)
y = np.random.randint(0,10,100) + x # agregamos tendencia
````

````python
x, y
````

## Grafico de lineas
```python
plt.plot(x, y)
```

````python
plt.plot(x, y)
plt.title('Lineas')
plt.xlabel('Tiempo')
plt.ylabel('Y')
plt.show()
````

#### Conocer los estilos disponibles en matplotlib

````python
plt.style.available
````

````python
plt.style.use('grayscale')
z = np.random.normal(0,1,100)
plt.plot(x, z)
plt.title('Nuevo Estilo: grayscale')
plt.show()
````

````python
plt.style.use('seaborn-bright')
plt.plot(x, z)
plt.title('Nuevo Estilo: seaborn-bright')
plt.show()
````

El anterior estilo está mal aplicado, debemos primero recetear los estilos antes de aplicar uno nuevo, se hace así:

````python
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
````

````python
plt.style.use('seaborn-bright')
plt.plot(x, z)
plt.xlabel("Tiempo")
plt.ylabel("Fuerza")
plt.title('Nuevo Estilo: seaborn-bright')
plt.show()
````
