## Gr치fica de pastel
usar el ultimo registro don la poblaci칩n de vaias zonas del mundo para realizar un gr치fico de pastel

```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.ticker import AutoMinorLocator, FuncFormatter

%matplotlib inline

df = pd.read_csv("DB/original.csv", sep=";", index_col=0)
df.tail() # ultimos cinco registros
```

```python
zonas = ['Latam', 'Europa', 'USA']

df['Otros'] = df['Mundo'] - df[zonas].sum(axis=1)
df.head()
```

```python
df.tail()
```

Ultimo A침o

```python
zonas.append('Otros')
ultimaFila = df[zonas].iloc[-1]
ultimaFila
```

## Gr치fica

```python
fig, ax = plt.subplots(figsize=(14,14))

colores = ['#63B8FF', '#EF5455', '#FAC744', '#8BD8BD'] # Colores, puedo usar RGB u HEX, el primer color en la lista, corresponde al primer color en la lista zonas

expTupla = (0.0, 0.0, 0.2, 0.0) #entre mas alto sea el valor despues de del ., mas se separa la porci칩n
ax.pie(ultimaFila, labels=zonas, textprops={'fontsize':15, 'fontname':'Serif', 'weight':'bold'},
       colors=colores, shadow=True, explode=expTupla, autopct='%.2f%%', pctdistance=0.8,
       startangle = 150, wedgeprops = {'linewidth':3}
      )
fig.suptitle("Distribuci칩n de Poblaci칩n", fontsize=25, fontname='Serif')
ax.set_title('A침o {}'.format(df.index[-1]), fontsize=25, fontname='Serif')
plt.show()
```

## Extrayendo las porciones
para esto se usa el침 atributo **explode**, elc aual recibe una tupla

## Agregar Etiquetas Porcentuales
- autopct = '%.2f%%'
- ptcdistance => recibe valor entre 0 y 1, modifica la posici칩n de la etiqueta

## Otros
- wedgeprops => grosor de la linea de la sombra
- startangle => angulo de inclinaci칩n de la grafica
