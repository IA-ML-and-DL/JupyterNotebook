```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

peliculas = pd.read_csv('DB/peliculas.csv', sep=';')
peliculas.head()
```

```python
peliculas.columns
```

```python
# creamos nuevo data frame solo con las columnas que necesitamos
datos = peliculas[['total_fb_likes', 'imdb_calificacion']]
datos.head()
```

```python
datos.describe()
```

```python
plt.style.use('seaborn-whitegrid')

# figura
fig, ax = plt.subplots(figsize=(15,8))

# Grafica
'''
    markers
    https://matplotlib.org/3.1.3/api/markers_api.html
    ".", ",", "o", "v", "^", "<", ">", "1", "2", "3", "4", "8", "s", "p", "P", "*", "h", "H", "+", "x", "X", "D", "d", "|", "_"
'''
ax.plot(datos.total_fb_likes, datos.imdb_calificacion, marker='o', linestyle='', markersize=9,
       color='#F0A07C', markeredgecolor='#4A274F', markeredgewidth=1.5)

# Etiquetas
ax.set_title('Crticas y Popularidad', fontname='fantasy', fontsize=25, va='bottom')
ax.set_xlabel('Lake in Facebook', fontname='fantasy', fontsize=20)
ax.set_ylabel('Calificacin en IMDB', fontname='fantasy', fontsize=20)

# Ticks
def formato(x, pos):
    return '%.0f\nmil likes'%(x/1000)

ax.set_xticklabels(list(ax.get_xticks()), fontname='fantasy', fontsize=13)
ax.xaxis.set_major_formatter(FuncFormatter(formato))

ax.set_yticklabels(list(ax.get_yticks()), fontname='fantasy', fontsize=13)

plt.show()
```

## 33. Graficando con scatter

Tomaremos las peliculas con duracin mayor a 150 minutos, usaremos un filtro

```python
peliculas2 = peliculas.loc[peliculas.duracion >= 150]
datos = peliculas2[['total_fb_likes', 'imdb_calificacion', 'presupuesto']]
datos
```

### Reestablercer los estylos

```python
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
plt.style.use('bmh')

# Figura
fig2, ax2 = plt.subplots(figsize=(15, 8))

# grafica
ax2.scatter(datos.total_fb_likes, datos.imdb_calificacion, marker='o', s=datos.presupuesto/100000, c='#2F3C7E'
           , edgecolor='#FBEAEB', linewidth=1.5, alpha=0.65)
# los atributos son diferentes segun el tipo de grafica
# s de size, puede ser un valor entero ola referencia una lista
# Etiquetas
ax2.set_title('Crticas y Popularidad', fontname='fantasy', fontsize=25, va='bottom')
ax2.set_xlabel('Likes in Facebook', fontname='fantasy', fontsize=20)
ax2.set_ylabel('Calificacin en IMDB', fontname='fantasy', fontsize=20)


plt.show()
```
