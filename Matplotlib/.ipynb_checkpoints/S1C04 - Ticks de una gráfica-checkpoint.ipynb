```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import math as m

from matplotlib.ticker import MultipleLocator, AutoMinorLocator
from matplotlib.patches import Circle
```

## Anatomia de una Grafica
#### Figura, axes y axis

```python
# Reiniciamos mpl y sembramos la semilla para los numeros aleatorios
mpl.rcParams.update(mpl.rcParamsDefault)
np.random.seed(123)
```

```python
x = np.linspace(0.5, 3.5, 100) # secuencia
y1 = 3 + np.cos(x)
y1
```

```python
y2 = 1 + np.cos(1 + x/0.75)/2
y2
```

```python
y3 = np.random.uniform(y1, y2, len(x)) # distrubici√≥n uniforme
y3
```

## Figura (contenedor)

```python
fig, ax = plt.subplots(figsize=(8,8)); # esto crea figura contenedor y axes, con 8 de ancho y 8 de alto
```

```python
fig
```

```python
ax
```

```python
ax.xaxis.set_major_locator(MultipleLocator(1.0000)) # los Ticks mayores seran multiplos de 1
```

```python
fig
```

```python
ax.xaxis.set_minor_locator(AutoMinorLocator(4)) # 4 ticks menores equidistantes
```

```python
fig
```

Para el eje y seria:

```python
ax.yaxis.set_major_locator(MultipleLocator(1.000))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))
fig
```

###  Circulos

```python


circulo = Circle((.5, .5, ), 0.15, linewidth=1, edgecolor='black', facecolor=(0,0,0,0.02)) # CoordenadaX, CoordenadaY, radio, grosor linea, bordecirculo
# facecolor(0,0,0,0.02) para crear el relleno transparente, se usa el 4 parametro en el RGB, el cual es la trasnsparencia, entre mas cercado a cero, mas transparente
ax.add_artist(circulo)
fig
```
