## 6. Funciones y m√©todos para una gr√°fica

```python
def addCircle(x, y):
    c1 = Circle((x, y), 0.15, linewidth=1, edgecolor='black', facecolor=(0,0,0,0.02))
    ax.add_artist(c1)
```

```python
def addText(x, y, text):
    ax.text(x, y, text, ha='center', weight='bold', color='blue', backgroundcolor='white', va = 'top')
```

```python
addCircle(0.5, 0.5)
addText(0.5, 0.3, 'Axis')
fig
```

```python
ax.tick_params(which='major', width= 1.0, length=10) # se notas mas los mayores y menores
ax.tick_params(which='minor', width= 1.0, length=10)
fig
```

### Colocar valores a los ticks menores

```python
def tickMenor(x, pos):
    if not x % 1.0:
        return ""
    else:
        return "%.2f"%x
    
ax.xaxis.set_minor_formatter(FuncFormatter(tickMenor)) # recibe una funci√≥n
fig
```
