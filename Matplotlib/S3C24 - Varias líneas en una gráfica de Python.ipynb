```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patheffects as pa

acciones = pd.read_csv('DB/facebook_vs_apple.csv', sep =';')
acciones.head()
```

```python
acciones.dtypes
```

## Cambiamos a tipo Date 

```python
acciones['Date'] = pd.to_datetime(acciones['Date'], format='%d/%m/%Y')
acciones.head()
```

```python
acciones.dtypes
```

## Cambiamos el indice

```python
acciones.set_index('Date', inplace =  True) #inplace para que reemplace el objeto acciones
acciones.head()
```

```python
plt.style.use('ggplot')
```

```python
fig, ax = plt.subplots(figsize = (15,8))

ax.plot(acciones['facebook'], label='Facebook', lw=2, color='#EC4D37', path_effects=[pa.SimpleLineShadow(), pa.Normal()]) # Agregouna grafica
ax.plot(acciones['apple'], label='Apple', lw=2, color='#1D1B1B', linestyle="-.", path_effects=[pa.SimpleLineShadow(), pa.Normal()]) # agrego una segunda grafica

# Leyenda
ax.legend(shadow= True, prop = {'family': 'Serif', 'size':15})

# Somra a la linea
# path_effects=[pa.SimpleLineShadow(), pa.Normal()]

#Etiquetas
ax.set_title('Facebook vs Apple', fontname='Serif', fontsize=25)
ax.set_ylabel('Precio en USD', fontname='Serif', fontsize=15)
ax.set_xlabel('Tiempo', fontname='Serif', fontsize=15)

#Lineas
ax.hlines(acciones['apple'].max(), acciones.index[0], acciones.index[-1], linestyle=':', color='#1D1B1B')
ax.hlines(acciones['apple'].min(), acciones.index[0], acciones.index[-1], linestyle=':', color='#1D1B1B')

# ax.hlines(acciones['facebook'].min(), acciones.index[0], acciones.index[-1], linestyle='-', color='#EC4D37')
# ax.hlines(acciones['facebook'].max(), acciones.index[0], acciones.index[-1], linestyle='-' , color='#EC4D37')

#Rango
amplitud = acciones['apple'].max() - acciones['apple'].min()
ax.annotate('Amplitud %.2f'%(amplitud), xy= (acciones['apple'].idxmax(), acciones['apple'].min() + (amplitud/2)),
            ha='center', va='center', fontname='Serif', fontsize=14)

#Fechas
ax.annotate("", xy=(acciones['apple'].idxmax(), acciones['apple'].max()),
           xytext =(acciones['apple'].idxmax(), acciones['apple'].min() + (amplitud) /1.9),
           arrowprops = {"width":1, "headwidth":7, "connectionstyle":"arc3", 'color':'black'})

#flecha abajo
ax.annotate("", xy=(acciones['apple'].idxmax(), acciones['apple'].min()),
           xytext =(acciones['apple'].idxmax(), acciones['apple'].min() + (amplitud) /2.1),
           arrowprops = {"width":1, "headwidth":7, "connectionstyle":"arc3", 'color':'black'})


plt.show()
```
